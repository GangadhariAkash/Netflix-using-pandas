```python
# Convert date columns to datetime
df['Release Date'] = pd.to_datetime(df['Release Date'], dayfirst=True, errors='coerce')
df['Netflix Release Date'] = pd.to_datetime(df['Netflix Release Date'], dayfirst=True, errors='coerce')

# Extract year from both dates
df['Release Year'] = df['Release Date'].dt.year
df['Netflix Release Year'] = df['Netflix Release Date'].dt.year

# Filter only Movies and Netflix release within 2 years of actual release
popular_films = df[
    (df['Series or Movie'] == 'Movie') &
    ((df['Netflix Release Year'] - df['Release Year']) <= 2)
]

# Display result
popular_films[['Title', 'Release Year', 'Netflix Release Year']].head(10)
```

```python
print(df.columns.tolist())
```

```python
popular_canada = df[
    (df['Country Availability'].str.contains('Canada', na=False)) &
    (df['Series or Movie'] == 'Movie') &
    ((df['Netflix Release Date'].dt.year - df['Release Date'].dt.year).abs() <= 2)
]

popular_canada[['Title', 'Country Availability']].head(10)
```

```python
# Ensure Runtime is a string
df['Runtime'] = df['Runtime'].astype(str)

# Filter only movies
movies = df[df['Series or Movie'] == 'Movie']

# Filter those with Runtime containing 1h or more
movies_1hr_or_more = movies[movies['Runtime'].str.contains('1h|2h|3h|4h', na=False)]

# Display first 10 results
movies_1hr_or_more[['Title', 'Runtime']].head(10)

```

```python
print("Total movies with runtime ≥ 1 hour:", len(movies_1hr_or_more))
```

```python
# Step 1: Convert Languages to string (to safely use .str functions)
df['Languages'] = df['Languages'].astype(str)

# Step 2: Filter TV Series that have more than one language (i.e., comma in 'Languages')
multi_lang_series = df[
    (df['Series or Movie'] == 'Series') &
    (df['Languages'].str.contains(',', na=False))
]

# Step 3: Display first 10
multi_lang_series[['Title', 'Languages']].head(10)
```

```python
# Step 1: Convert necessary columns to string (to avoid AttributeError)
df['Country Availability'] = df['Country Availability'].astype(str)
df['Genre'] = df['Genre'].astype(str)

# Step 2: Filter Movies available in Japan with multiple genres
japan_multi_genre_movies = df[
    (df['Series or Movie'] == 'Movie') &
    (df['Country Availability'].str.contains('Japan', na=False)) &
    (df['Genre'].str.contains(',', na=False))  # Comma indicates multiple genres
]

# Step 3: Display the result
japan_multi_genre_movies[['Title', 'Genre', 'Country Availability']].head(10)
```

```python
# Step 1: Convert scores to numeric (handle errors safely)
df['Hidden Gem Score'] = pd.to_numeric(df['Hidden Gem Score'], errors='coerce')
df['IMDb Score'] = pd.to_numeric(df['IMDb Score'], errors='coerce')

# Step 2: Filter only Movies
movies_df = df[df['Series or Movie'] == 'Movie']

# Step 3: Compute Total Rating
movies_df['Total Rating'] = movies_df['Hidden Gem Score'] + movies_df['IMDb Score']

# Step 4: Filter movies with total rating >= 15
highly_rated_movies = movies_df[movies_df['Total Rating'] >= 15]

# Step 5: Display result
highly_rated_movies[['Title', 'Hidden Gem Score', 'IMDb Score', 'Total Rating']].head(10)
```

```python
# Step 1: Convert necessary columns to string
df['Genre'] = df['Genre'].astype(str)
df['Country Availability'] = df['Country Availability'].astype(str)

# Step 2: Filter based on all conditions
dramatic_japan_series = df[
    (df['Series or Movie'] == 'Series') &
    (df['Genre'].str.contains('Drama', case=False, na=False)) &
    (df['Country Availability'].str.contains('Japan', na=False))
]

# Step 3: Display the result
dramatic_japan_series[['Title', 'Genre', 'Country Availability']].head(10)
```

```python
print("Number of dramatic TV series from Japan:", len(dramatic_japan_series))
```

```python
# Step 1: Convert 'Director' to string to avoid errors
df['Director'] = df['Director'].astype(str)

# Step 2: Filter movies directed by a person whose name starts with 'C'
directed_by_c = df[
    (df['Series or Movie'] == 'Movie') &
    (df['Director'].str.startswith('C', na=False))
]

# Step 3: Display result
directed_by_c[['Title', 'Director']].head(10)
```

```python
print("Number of films directed by a person whose name starts with 'C':", len(directed_by_c))
```

```python
# Step 1: Ensure 'Title' and 'Director' columns are strings
df['Title'] = df['Title'].astype(str)
df['Director'] = df['Director'].astype(str)

# Step 2: Filter based on conditions
nolan_o_movies = df[
    (df['Series or Movie'] == 'Movie') &
    (df['Director'].str.strip().str.lower() == 'christopher nolan') &
    (df['Title'].str.endswith('o'))
]

# Step 3: Display results
nolan_o_movies[['Title', 'Director']].head(10)
```

```python
print("Number of Nolan-directed films ending with 'o':", len(nolan_o_movies))
```

```python
# Step 1: Convert columns to string to avoid type issues
df['Director'] = df['Director'].astype(str)
df['Writer'] = df['Writer'].astype(str)

# Step 2: Filter where both Director and Writer are Christopher Nolan
nolan_directed_written = df[
    (df['Series or Movie'] == 'Movie') &
    (df['Director'].str.strip().str.lower() == 'christopher nolan') &
    (df['Writer'].str.strip().str.lower() == 'christopher nolan')
]

# Step 3: Display result
nolan_directed_written[['Title', 'Director', 'Writer']].head(10)
```

```python
print("Number of films directed and written by Christopher Nolan:", len(nolan_directed_written))
```
